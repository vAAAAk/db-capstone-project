```python
# Customers with 60$ in payment for promotional campaign
try:
    tmp_sql = """
    SELECT 
        c.customer_id,
        c.customer_name,
        c.customer_phone,
        c.customer_email
    FROM 
        customers AS c
    WHERE c.customer_id IN (
        SELECT 
            o.customers_customer_id as customer_id
        FROM
            orders AS o
        GROUP BY o.customers_customer_id
        HAVING SUM(o.total_cost) > 60
        );
    """
    cursor.execute(tmp_sql)
    result = cursor.fetchall()
    if not result:
        print("No customers found with more than $60 in payments.")
    else:
        print("Customers with more than $60 in payments:")
        for row in result:
            print(row)
except Error as e:
    print("Error executing promotional campaign query")
    print("Error Code :", e.errno)
    print("Error Message :", e.msg)
```
